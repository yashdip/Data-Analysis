# Studying Survivers and deaths by age group and sex

```python
data['AgeGroups']=pd.cut(data['Age'],bins=3,labels=('young','middle_age','old'))
```

```python
data.groupby(['AgeGroups','Sex'],as_index=False)['Survived'].sum()
```

```python
# percentage of males and females survived
len(data[(data['Sex']=='male') & (data['Survived']==1)])*100/len(data['Sex']=='male')
```

```python
len(data[(data['Sex']=='female') & (data['Survived']==1)])*100/len(data['Sex']=='female')
```

```python
def encodesex(x):
    if x=='male':
        return 1
    else:
        return 0
```

```python
data['En_Sex']=data['Sex'].apply(lambda x: encodesex(x))
```

```python
data.groupby(['AgeGroups','Survived'])['En_Sex'].mean()
```
